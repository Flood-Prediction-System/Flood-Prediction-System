```python
selectives= ['MonsoonIntensity',
    'TopographyDrainage',
    'RiverManagement',
    'Deforestation',
    'Urbanization',
    'ClimateChange',
    'DamsQuality',
    'Siltation',
    'Encroachments',
    'DrainageSystems',
    'Landslides',
    'Watersheds',
    'DeterioratingInfrastructure',
    'WetlandLoss',
    'InadequatePlanning',
    'FloodProbability']

corr_matrix = df[selectives].corr()

# plot heatmap
plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")

plt.xticks(rotation=45, ha='right')
plt.title("Correlation Heatmap (Selected Features)")
plt.show()
```

```python
for i in selectives:
    plt.figure()
    plt.boxplot(df[i])
    plt.xlabel(i)
    plt.ylabel("FREQUENCY")
    plt.title("DETECTING OUTLIERS IN "+i)
    plt.show()
```

```python
df2= df[selectives]
df2
```

```python
df2.shape
```

```python
import numpy as np

features_cap = np.array([
    'TopographyDrainage',
    'RiverManagement',
    'DamsQuality',
    'DrainageSystems',
    'DeterioratingInfrastructure',
    'Watersheds',
    'WetlandLoss',
    'Encroachments',
    'InadequatePlanning'
])
```

```python
def bounds(f):
    q1=df2[f].quantile(0.25)
    q3=df2[f].quantile(0.75)
    iqr=q3-q1
    lower= q1-1.5*(iqr)
    upper= q3+1.5*(iqr)
    df2[f]=df2[f].clip(lower,upper)

for i in features_cap:
    bounds(i)
    print("OUTLIERS CAPPED")
print(df2)
```

```python
for i in selectives:
    plt.figure()
    plt.boxplot(df2[i])
    plt.xlabel(i)
    plt.ylabel("FREQUENCY")
    plt.title("DETECTING OUTLIERS IN "+i)
    plt.show()
```
